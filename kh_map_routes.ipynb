```python
import pandas as pd
import folium
```

```python
trips = pd.read_csv('./sch/shapes.txt')
trips.head()
```

```python
trip_1 = trips.loc[trips['shape_id'] == 275145]
trip_1 = trip_1.sort_values(by=['shape_pt_sequence'])
```

```python
trip_2 = trips.loc[trips['shape_id'] == 275075]
trip_2 = trip_2.sort_values(by=['shape_pt_sequence'])
```

```python
t1 = list(zip(trip_1.shape_pt_lat, trip_1.shape_pt_lon))
t2 = list(zip(trip_2.shape_pt_lat, trip_2.shape_pt_lon))
```

```python
map = folium.Map(location=[40,-75.154839], zoom_start= 11)

folium.PolyLine(t1, color='red', weight=10, opacity=0.8).add_to(map)
folium.PolyLine(t2, color='blue', weight=10, opacity=0.8).add_to(map)

map.save(outfile='./maps2/septa_trips.html')
```
